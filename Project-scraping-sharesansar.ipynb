```python
from bs4 import BeautifulSoup
import requests #we use this 'requests' package as this sharesanasr live trading might be difficult for Beautiful Soup to parse; the other reason is because we don't have a html file, when we have an URL.
url = BeautifulSoup("https://www.sharesansar.com/live-trading", 'html.parser') #url is a variable 
soup = requests.get(url)
soup
```

```python
code = BeautifulSoup(soup.text,"lxml")
code
```

```python
table_code = code.table
table_code
```

```python
tags = table_code.find_all('tr')
tags
```

```python
data = tags[0].text #converting into text
data
```

```python
for i in tags:
    data = i.text
    print(data)
```

```python
#now put this into a list as the previous output has lots of spaces in between
list_data = []
for i in tags:
    data = i.text
    list_data.append(data)

print(list_data)
```

```python
for i in list_data:
    print(i.replace("                                        ","")) #Replacing the void or white space in the previous output
```

```python
cleaned_data = []
for i in list_data:
    x = i.replace("                                        ","")
    cleaned_data.append(x)
cleaned_data
```

```python
#This output is in string, that needs to be converted into a list format; remove \n that appears at the output
cleaned_data = []
for i in list_data[1:]:
    x = i.replace("                                        ","")
    x = x.split('\n\n')
    cleaned_data.append(x)
cleaned_data
```

```python
#the previous output has \n in some rows, that needs to be removed
cleaned_data = []
for i in list_data[1:]:
    x = i.replace("                                        ","")
    x = x.split('\n\n')
    x = [j.replace('\n','') for j in x]
    cleaned_data.append(x)
    
cleaned_data
```

```python
#now we need to remove the '' that appears on each row output
#the previous output has \n in some rows, that needs to be removed
cleaned_data = []
for i in list_data[1:]:
    x = i.replace("                                        ","")
    x = x.split('\n\n')
    x = [j.replace('\n','') for j in x]
    cleaned_data.append(x[1:])
    
cleaned_data
```

```python
#Now need to append these output to a dataframe
#we have just completed doing the row only, now need to do the column
```

```python
col = list_data[0].split('\n')[1:-1]
col
```

```python
#now need to bring row and column together

import pandas as pd
df = pd.DataFrame(cleaned_data,columns = col)
df
```

```python
#now the numbers showing in the table are in string format not in float
#we first need to clean the comma appearing in the numbers
#converting string into float
#run these following codes only once otherwise there might be some errors
df['LTP'] = [float(i.replace(',','')) for i in df['LTP']] 
df['Point Change'] = [float(i) for i in df['Point Change']]
df['% Change'] = [float(i) for i in df['% Change']]
df['Open'] = [float(i.replace(',','')) for i in df['Open']]
df['High'] = [float(i.replace(',','')) for i in df['High']]
df['Low'] = [float(i.replace(',','')) for i in df['Low']]
df['Volume'] = [float(i.replace(',','')) for i in df['Volume']]
df['Prev. Close'] = [float(i.replace(',','')) for i in df['Prev. Close']]
```

```python
#The above error came beacuse I ran the code twice. It's not a problematic as the data has been converted to float from the string and also the comma appearing were removed
# Now we can do the data visuationisation on these output
```
